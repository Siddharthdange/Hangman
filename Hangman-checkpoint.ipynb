```python
#Main function

def game(correct):
    i=0
    
    wordlen = len(correct)
    guess = "*"*wordlen
    guessed = []
    n = 0
    
    #For loop
    for x in range(0,26):
        print("Guess the value")
        gl = input()
        
        #If already guessed
        if gl in guessed:
            print("Already guessed")
            guessed.append(gl)
            
            #If correct
        elif gl in correct:
            print("Correct guess")
            guessed.append(gl)
            for i in range(wordlen):
                if gl == correct[i]:
                    guess = guess[:i]+correct[i]+guess[i+1:]
                    print(guess)
                    i+=1 
                    
        #If not correct            
        else:
            print("Oops not correct")
            guessed.append(gl)
            if(n<=7):
                print(hanger[n])
                n+=1
            else:
                break
        if guess == correct:
            break
            
    #Checking final output        
    if guess == correct:
        print("Yeah congo")
        print(hanger[8])
        exit()
    else:
        print("Sorry you lost")
        print(hanger[9])
```

```python
#Driver Code

print("Welcome")

#Datasets
fruits = ["apple","mango","guava","banana","orange","litchi","cherry","dragonfruit","kiwi","papaya","pomegranate","pineapple","coconut","pear","plum","grapes","strawberry","watermelon","muskmelon","raspberry","blackberry"]
colors = ["red","yellow","green","blue","purple","violet","pink","white","brown","black","cream","peach","grey","orange","magenta","maroon","silver","golden","mustard"]

#Importing random library
import random

s=1
while s == 1:
    print("Would u like to continue? Type 1 for yes.")
    s = int(input())
    
    #Asking category
    if s==1:
        print("Choose Category: Type 1 for Fruits and 2 for Colors ")
        choice = int(input())
        if choice == 1:
            correct = random.choice(fruits)
            print("*"*len(correct))
            game(correct)
        elif choice == 2:
            correct = random.choice(colors)
            print("*"*len(correct))
            game(correct)
        else:
            print("Invalid Choice..Start again")
    else:
        print("Thank You")
        exit()
```
